```python
#引入包
import pandas
import jieba
import re 

#打开文档
fb=open("C:\\Users\\LRX\\Desktop\\Python数据分析基础\\返校大作战.txt","r",encoding="utf-8")
text1=fb.read()
fb.close()
#print(text1)

#对文档预处理（去掉符号等）
pattern1 = re.compile(r"：|？|，|。|——|【|】|（|）|:|？|/|-|(|)|、|\n|\!|、|“|”") # 定义正则表达式匹配模式
text1=re.sub(pattern1,"",text1)#符合模式的符合去掉

#文本分词
text1=jieba.lcut(text1)
#print(text1)
#去除中文中常见的助词
remove_words = [u'的', u'，',u'和', u'是', u'随着', u'对于', u'对',u'等',u'能',u'都',u'。',u' ',u'、',u'中',u'在',u'了',
                u'通常',u'如果',u'我们',u'需要',u"是",u"了",u"的",u"…",u"吗"] # 自定义去除词库

for word in text1:
    if word in remove_words:
        text1.remove(word)
print(text1)
```

```python
#生成词频统计

data={}
for word in text1:
    if len(word)>1:
        data[word]=data.get(word ,0)+1
#data=sorted(data,key =lambda x:x[1],reverse=True)
print(data)
```

```python
#分离词和词频

import pandas as pd
pd.set_option("display.max_rows",None)
word_list=list(data.keys())
#print(word_list)
frequency_list=list(data.values())
#print(frequency_list)

from pandas.core.frame import DataFrame
#把词和词频对应生成矩阵
result={"word":word_list,"frequency":frequency_list}
pd_data=DataFrame(result)
print(pd_data)
```
