```python
def matrixMul(A,B):#矩阵实现斐波那契数列
    line_A,line_B,column_B=len(A),len(B),len(B[0])
    res = [[0] * column_B for i in range(line_A)]
    for i in range(line_A):
        for j in range(column_B):
            for k in range(line_B):
                res[i][j] += A[i][k]*B[k][j]
    return res

def fibonacci(n):
    
    if n <=1:
        return max(n,1)
    res = [[1,1],[0,1]]#单位矩阵
    A = [[1,1],[1,0]]
    while n:
        if n & 1:
            res = matrixMul(res,A)
        A = matrixMul(A , A)
        n >>= 1
    return res[0][1]
```

```python
def mul(num1,num2):#大数字乘法结果数列显示
    if type(num1)==int:
        n1=[int(i) for i in str(num1)]
    else:
        n1=num1
    if type(num2)==int:
        n2=[int(i) for i in str(num2)]
    else:
        n2=num2
    n1.reverse()
    n2.reverse()
    n3=[]
    for i0 in range(len(n1)+len(n2)):
        n3.append(0)
    for i1 in range(len(n1)):
        for i2 in range(len(n2)):
            n3[i1+i2] += n1[i1]*n2[i2]
    for i3 in range(len(n3)):
        if(n3[i3]>9):
            n3[i3+1] += int(n3[i3]/10)
            n3[i3] = n3[i3]%10
    n3.reverse()
    return n3
def Fmul(n):
    x = [1]
    for i in range(1,n):
        x = mul(x,fibonacci(i))
    return x
def adopt(x):
    m = 0 
    anw=""
    for i in range(len(x)):
        if(x[i] == 0):
          m=m+1  
        else:
            break
    del x[:m]
    for i1 in range(len(x)):
        anw += str(x[i1])
    return anw

def FMUL(n):
    n = Fmul(n)
    n = adopt(n)
    return n
```

```python
def Mul(n):#基础乘法
    x = 1
    for i in range(1,n):
        x = x*fibonacci(i)
    return x
```
