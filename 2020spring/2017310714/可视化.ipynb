```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
#解决中文显示问题
from pylab import *  
mpl.rcParams['font.sans-serif'] = ['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  #用来正常显示负号
```

```python
mask = pd.read_csv('mask.csv', encoding='gbk')   
mask.info()
```

```python
fig,axes=plt.subplots(1,3)      
plt.subplots_adjust(wspace=0.4) 
sns.distplot(mask['月销量'],rug=True,ax=axes[0],hist=False) 
sns.distplot(mask['价格'],rug=True,ax=axes[1],kde=False)   
plt.hist(mask['评价数'],bins=30,density=True,alpha=0.4)
plt.show()
```

```python
plt.figure(figsize=(12,5))
data1=mask[mask['适合肤质'] != '任何肤质']
sns.countplot('适合肤质',data=data1,order = data1['适合肤质'].value_counts().index,palette='Blues_r')
```

```python
sns.boxenplot(x='适合肤质',y=log(data1['月销量']),data=data1,palette="Set3")
plt.ylabel('log(月销量)')
```

```python
plt.figure(figsize=(12,5))
data2=mask[mask['产地'] != '中国']
sns.countplot('产地',data=data2,order = data2['产地'].value_counts().index,palette='Reds_r')
```

```python
sns.pairplot(mask[['月销量','价格','适合肤质']],hue='适合肤质')
```
