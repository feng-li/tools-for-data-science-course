```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
html=urlopen('https://feng.li/python/')

bs = BeautifulSoup(html.read(),'html.parser')
print(bs)
```

```python
divlist= bs.find_all('div',{'class':'entry-content'})
```

```python
for div in divlist:
    print(div.get_text())
```
