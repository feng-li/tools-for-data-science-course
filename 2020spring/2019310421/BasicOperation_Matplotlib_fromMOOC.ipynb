### 画一条正弦曲线

```python
import  numpy as np 
import matplotlib.pyplot as plt
x = np.arange(0,20,0.5)  #   arange表示范围。 0-20的范围，步长为0.5
y = np.sin(x)           # 计算函数值
plt.plot(x,y,'bp-')  #  画图
plt.show()
```

### 在一个图同时画正弦 和 余弦曲线

```python
import numpy as np 
import matplotlib.pyplot as plt
x = np.arange(0,10,0.2)
y = np.sin(x)     #计算值
z = np.cos(x)     #计算值
plt.plot(x,y,'r:',label='sine')    #  label 是用来设置图例的  
plt.plot(x,z,'g-',label='cosine')
plt.title('sine and cosine')         #设置图的名字
plt.xlabel('X')               # 设置 x轴  的名字
plt.legend()          #搭配 label使用  【不用legend的话，不显示图例】
```

### 正态分布的直方图

```python
import numpy as np
import matplotlib.pyplot as plt
mu = 120
sigma = 15
x = mu + sigma*np.random.randn(1000)
plt.hist(x,50,density=True,facecolor = 'g',alpha=0.3)#hist：直方图 alpha：透明度
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60,.025,r'$\mu = 100,\ \sigma=15$')# 添加文本内容，前两个参数用来定位
plt.xlim(40,160)          #限定横纵坐标的范围
plt.ylim(0,.04)
plt.grid(True)          # 添加网格
```
