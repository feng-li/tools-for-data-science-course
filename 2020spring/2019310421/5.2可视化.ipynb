```python
from numpy import median

fig= plt.figure(figsize=(16,16))
ax= fig.add_subplot(3,2,1)
sns.barplot(x="产地",y="价格",data=mask,estimator=mean,
            ci=75,order=['美国','韩国','英国','日本',
                         '其他','中国'],capsize=0.05)
plt.title('指定order和capsize =0.05')

ax= fig.add_subplot(322)
sns.barplot(x="产地",y="月销量",
            data=mask,estimator=median,ci=20,color="salmon")
plt.xticks(rotation=-30)  #旋转标签
plt.title('指定color')

ax=fig.add_subplot(323)
sns.barplot(x="产地", y="月销量",hue="适合肤质",
            data=mask, estimator=median,ci=20)
plt.title('指定hue')
plt.legend(loc='upper right')

ax=fig.add_subplot(324)
sns.barplot(x="产地", y="月销量", data=mask,
            estimator=median,ci=20,color="salmon",saturation=1.2)
plt.title('指定saturation')

ax=fig.add_subplot(3,2,5)
bardata = mask[['产地',
       '月销量']].groupby(['产地']).median().sort_values(by='月销量',
                               ascending = False).reset_index()
g = sns.barplot(x="产地", y="月销量",data=bardata, palette="Paired")
for index,row in bardata.iterrows():
    g.text(row.name,row.月销量,int(row.月销量),
           color="black",ha="center") #ha调整位置
plt.title('指定palette和标签')

ax=fig.add_subplot(3,2,6)
g = sns.barplot(x="月销量", y="产地",data=bardata, palette="Paired",
                orient = "h")
for index,row in bardata.iterrows():
    g.text(row.月销量,row.name,int(row.月销量),color="black",
           va="center") #va调整位置
plt.title('指定orient和标签')##注意改变x和y
```

# 词云图  爬取中文文本数据

```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get('http://www.un.org/zh/charter-united-nations/')
soup = BeautifulSoup(page.content, 'html.parser')  #解析网页
Link=soup.find_all("ul")[9].find_all("a")  #获取每条宪章的链接
UN = 'http://www.un.org/zh/'
hr=list()
for link in Link:
    hr.append(UN+link.get('href'))  #拼接链接
for i in range(len(hr)):
    hr[i]=hr[i].replace('../','')  #链接中的多余字符去掉
#爬取每条宪章的网页写入txt
with open("UNCharter.txt", "w",encoding='utf-8') as text_file:
    for k in range(len(hr)):
        r = requests.get(hr[k])
        sp = BeautifulSoup(r.content, 'html.parser')
        s_div=sp.find_all("section",class_="col-sm-9")[0]
        s_p=s_div.find_all({"h3","p"})
        for i in s_p:
            text_file.write(''.join(i.strings)+'\n')   
```

## 分词画词云图

```python
import jieba
from wordcloud import WordCloud, STOPWORDS      #STOPWORDS为英文停用词
import matplotlib.pyplot as plt                                                  
import numpy as np
from imageio import imread                        #读取图像
```

```python
 使用jieba分词，获取词的列表,再用空格分隔拼接
txt2 = ' '.join(jieba.lcut(txt1))  
#设置词云图对象属性
WC = WordCloud(font_path = 'C:\\Windows\\Fonts\\simhei.TTF',             #设置字体(中文必须设置)
               max_words=200,mask = imread('heart.jpeg',pilmode="RGB"),  #mask设置形状
               height= 1000,width=1000,collocations=False,              #collocations避免重复词
               background_color='white',mode='RGBA') 
con = WC.generate(txt2)
con.to_file("ciyun.png")  #导出图片
plt.imshow(con)
plt.axis("off")  #不显示坐标轴
```
