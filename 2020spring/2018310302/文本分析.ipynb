```python
#读入上次抓取的新冠病毒主题文本
import csv
sentence=""
with open('test1.csv','r')as f:
    f_csv = csv.reader(f)
    for row in f_csv:
        for i in row[3:4]:
            sentence+=i
print(sentence)
```

```python
#使用jieba库进行关键词提取
import jieba.analyse
jieba.analyse.extract_tags(sentence, topK=15, withWeight=False, allowPOS=())
```

```python
jieba.analyse.textrank(sentence, topK=15, withWeight=False, 
                       allowPOS=('ns', 'n', 'vn', 'v'))
```

```python
#分词后进行词频统计
from collections import Counter
import jieba

# 全模式
cut_words=""
seg_list = jieba.cut(sentence, cut_all=True)
cut_words+=(" ".join(seg_list))
all_words=cut_words.split()
print(all_words)
print("================================================")
c=Counter()
for x in all_words:
    if len(x)>1 and x != '\r\n':
        c[x] += 1

print('\n词频统计结果：')
for (k,v) in c.most_common(10):# 输出词频最高的前两个词
    print("%s:%d"%(k,v))  
```

```python
with open('test2.csv','w')as f:
    f_csv = csv.writer(f)
    for (k,v) in c.most_common():
        row=[k,v]
        f_csv.writerow(row) #记录词频矩阵
```

```python
import matplotlib.pyplot as plt 


def plot_resumes(plt):
    data = [ ("big data", 100, 15), ("Hadoop", 95, 25), ("Python", 75, 50),
         ("R", 50, 40), ("machine learning", 80, 20), ("statistics", 20, 60),
         ("data science", 60, 70), ("analytics", 90, 3),
         ("team player", 85, 85), ("dynamic", 2, 90), ("synergies", 70, 0),
         ("actionable insights", 40, 30), ("think out of the box", 45, 10),
         ("self-starter", 30, 50), ("customer focus", 65, 15),
         ("thought leadership", 35, 35)]

    def text_size(total):
        """equals 8 if total is 0, 28 if total is 200"""
        return 8 + total / 200 * 20

    for word, job_popularity, resume_popularity in data:
        plt.text(job_popularity, resume_popularity, word,
                 ha='center', va='center',
                 size=text_size(job_popularity + resume_popularity))
    plt.xlabel("Popularity on Job Postings")
    plt.ylabel("Popularity on Resumes")
    plt.axis([0, 100, 0, 100])
    plt.show() 
    
plot_resumes(plt)
```
