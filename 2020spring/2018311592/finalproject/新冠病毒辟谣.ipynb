```python
import requests
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'}
```

```python
titles = []
dates = []
results = []
explains = []
tags = []
for i in range(61):
    url = 'https://vp.fact.qq.com/loadmore?artnum=0&page='+str(i)
    print('第%d页面抓取'%i)
    dfs = requests.get(url,headers = headers).json()['content']
    for df in dfs:
        titles.append(df['title'])
        dates.append(df['date'])
        results.append(df['result'])
        explains.append(df['explain'])
        tags.append(df['tag'])
```

```python
import pandas as pd
datas = pd.DataFrame()
datas['titles'] = titles
datas['date'] = dates
datas['result'] = results
datas['explain'] = explains
datas['tag'] = tags
```

```python
datas.to_excel('D:/腾讯较真.xlsx',index = None)
```

```python
import pyecharts.options as opts
from pyecharts.charts import Line,Pie,Bar

def line_base(x,y):
    c = (
        Line()
        .add_xaxis(x)
        .add_yaxis("",y)
        .set_global_opts(title_opts=opts.TitleOpts(title="每日谣言数"))
    )
    return c
```

```python
df1 = datas.groupby('date').size()
x = list(df1.index)
y = [float(d) for d in list(df1.values)]
c1 = line_base(x,y)
c1.render_notebook()
```

```python
df2 = datas.groupby('explain').size()
x = list(df2.index)
y = [float(d) for d in list(df2.values)]
```

```python
c = (
        Pie()
        .add(
            "",
            [list(z) for z in zip(x,y)],
            radius=["40%", "55%"],#圆环粗细
            #rosetype="radius",
            #设置标签
            label_opts=opts.LabelOpts(
                position="outside", #显示在外
                formatter="{a|{a}}{abg|}\n{hr|}\n {b|{b}: }{c}  {per|{d}%}  ",
                background_color="#eee",#背景色
                border_color="#aaa",#边框颜色
                border_width=1,#边框线的宽度
                border_radius=4,#边框形状
                rich={
                    #设置series
                    "a": {"color": "#999", "lineHeight": 22, "align": "center"},
                    #画线
                    "hr": {
                        "borderColor": "#aaa",
                        "width": "100%",
                        "borderWidth": 0.5,
                        "height": 0,
                    },
                    #添加黑色底框
                    "per": {
                        "color": "#eee",
                        "backgroundColor": "#334455",
                        "padding": [2, 3],  
                    },
                },
            ),
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="谣言是否属实占比"),
            legend_opts=opts.LegendOpts(
                type_="scroll", pos_left="83%", orient="vertical"
            ),
        )
    )
c.render_notebook()
```

```python
lists = sum(list(datas['tag']),[])
```

```python
from collections import Counter
df3 = pd.DataFrame([dict(Counter(lists))]).T
df3.columns = ['count']
df3 = df3.sort_values(by = 'count',ascending = False).reset_index()
df3 = df3[df3['index'].map(lambda x:len(x) > 0)]
```

```python
df3
```

```python
def bar_base(x,y):
    c = (
        Bar()
        .add_xaxis(x)
        .add_yaxis("",y,color = 'skyblue')
        .set_global_opts(title_opts=opts.TitleOpts(title="谣言tag统计分布", subtitle=""))
        .set_series_opts(label_opts=opts.LabelOpts(position="right",is_show = False))
        .reversal_axis()
    )
    return c
```

```python
dff3 = df3.iloc[::-1]
x = list(dff3.iloc[-20:]['index'])
y = [float(d) for d in list(dff3.iloc[-20:]['count'])]
c2 = bar_base(x,y)
c2.render_notebook()
```
