```python
import word2vec
import gensim
```

### 导入cor数据进行模型训练

```python
 from gensim.models import word2vec
wd_sentences = word2vec.Text8Corpus("k.txt")  # 加载语料
wd_model = word2vec.Word2Vec(wd_sentences, size=200)  
```

### 输出各个词向量以及其矩阵

```python
print(wd_model.wv.index2word)
```

```python
for word in wd_model.wv.index2word:
    print(word,wd_model[word])  
```

### 相似度展示

```python
def getSimilarity(s):
    print('similarity by word of :'+s)
    for k in wd_model.similar_by_vector(s,5):
        print(k[0],k[1])
getSimilarity('肺炎')
```

### 总结
因为之前跑出来的词只有2800+个，导致整体的样本量较小。训练出来的相似程度都很小。

解决这一问题可以扩大样本容量。
