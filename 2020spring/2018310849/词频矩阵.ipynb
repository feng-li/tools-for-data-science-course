```python
import pandas as pd
import jieba
```

```python
data=pd.read_csv("新冠病毒.csv",encoding="GBK")
```

```python
data
```

```python
df=data["body"].astype(str)
stopwords_file = "stopwords.txt"
```

```python
f=open("test.txt", "r", encoding='UTF-8')
stopwords_list = [word.strip() for word in f.read()]
```

```python
def filter_stopwords(txt):

    """过滤停用词"""

    sent = jieba.lcut(txt)

    words = []

    for word in sent:

        if(word in stopwords_list):

            continue

        else:

            words.append(word)

    return words
```

```python
df = df.apply(filter_stopwords)

df.head()
```

```python
from nltk import FreqDist
```

```python
all_words = []

for comment in df:

    all_words.extend(comment)
```

```python
fdisk = FreqDist(all_words)

TOP_COMMON_WORDS = 1000

most_common_words = fdisk.most_common(TOP_COMMON_WORDS)
```

```python
most_common_words[:20]
```

```python
out=pd.DataFrame()
```

```python
for i in most_common_words:
    out[i[0]] =  [1 if i[0] in  x else 0 for x in df]
```
