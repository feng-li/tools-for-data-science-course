```python
with open("text.txt",encoding='UTF-8') as f:
    contents = f.read()
print("contents变量的类型：", type(contents))
print(contents)
```

清洗数据

```python
def handle_data(data):
    for i in range(10):
        data=data.replace(str(i),'')
    data=data.replace('\r','')
    data=data.replace('\n','')
    data=data.replace('\t','')
    data=re.sub("[%s]+" %punctuation, "",data)
    data=data.replace(' ','')
    return data
```

```python
correct=handle_data(contents)
```

自定义字典格式：  
邢青松 n  
全国人大代表 n  
固城湖 n  

```python
jieba.load_userdict('xing dict.txt')
```

分词

```python
fenci=jieba.lcut(correct,cut_all=False)
```

```python
for i in fenci:
    print(i,end='/')
```

```python
counts={}
```

```python
for word in fenci:
    counts[word] =fenci.count(word)
print(counts)
```

按照词频降序排列

```python
items = list(counts.items())
items.sort(key=lambda x: x[1], reverse=True)  # 默认升序，reverse=True为降序
print(dict(items))
```

```python
dic=dict(items)
for key,value in dic.items():
    print('{}:{}'.format(key,value))
```

将词频结果以csv格式写入Excel中

```python
with open('counts.csv', 'a', newline='')as f:
    write = csv.writer(f)
    write.writerow(["word", "count"])
    for i in range(len(items)):
        word, count = items[i]    # 返回相对应的键值对
        print(f"{word}:{count}")
        write.writerow([word, count])
```
